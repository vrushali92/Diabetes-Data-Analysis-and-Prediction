```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/diabetes_clean.csv")
data
```

```python
sns.countplot(data = data, x = "gender")
```

```python
sns.countplot(data = data, x = "hypertension", hue = "is_diabetic")
```

```python
sns.countplot(data = data, x = "heart_disease", hue = "is_diabetic")
```

```python
sns.countplot(data = data, x = "smoking_history", hue = "is_diabetic")
```

```python
sns.countplot(data = data, x = "is_diabetic")
```

```python
sns.boxplot(data = data, x = "age")
```

ðŸ’¡Insight: There are no outliers in age column.

```python
sns.histplot(data = data, x = "age", hue = "is_diabetic")
```

ðŸ’¡ Insight: Age is most important factor in causing diabetes. People of age 80 have highest number of diabetes.

```python
sns.boxplot(data = data, x = "bmi_score")
```

ðŸ’¡Insight: There are some outliers in bmi_score having values beyond 70.

```python
sns.boxplot(data = data, x = "haemoglobin_level")
```

ðŸ’¡Insight: There are 2 outliers in haemoglobin_level having values 8.9, 9.

```python
sns.displot(data = data, x = "haemoglobin_level", hue = "is_diabetic")
```

ðŸ’¡Insight: Looks like people have HbA1c_level between 6.5 and 9 have high chances of getting diabetes.

```python
sns.boxplot(data = data, x = "blood_glucose_level")
```

ðŸ’¡Insight: There are 3 outliers in blood_glucose_level having values between 250 and 300.

```python
sns.displot(data = data, x = "blood_glucose_level", hue = "is_diabetic")
```

ðŸ’¡Insight: People who have blood_glucose_level between 225 and 300 definitely have diabetes. 
    And there are also people who got diabetes having blood_glucose_level between 125 and 200.

```python
data.head()
```

```python
sns.scatterplot(data = data, x = "blood_glucose_level", y = "is_diabetic")
plt.title("Relation between blood_glucose_level and is_diabetic")
```

ðŸ’¡Insight: People having diabetes have glucose level beyond 125. But that is not the only factor in causing diabetes.
    There are people who have glucose level between 125 and 200 don't have diabetes. There might be other factors contributing to this situation.

```python
sns.scatterplot(data = data, x = "haemoglobin_level", y = "is_diabetic")
```

ðŸ’¡Insight: People having diabetes have haemoglobin_level beyond 5.5. But that is not the only factor in causing diabetes.
    There are people who have haemoglobin_level between 5.5 and 6.5 don't have diabetes. There might be other factors contributing to this situation.
